```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(30, 8), dpi=80, facecolor='w', edgecolor='k')
m.bar(df['Primary Genre'], df['Average User Rating'])
m.show()
print("Inference/Observation:\nThis plot shows a bar graph of all the primary genres available in the dataset against the respective ratings from users. The plot makes it clear that the primary genre such as Games, Entertainment, Sports, Refernce, Education, utilities, Productivity, News, Social Network, Food and Drink games have a maximum average user rating of 5.0 whereas the minimally rated ones are the medical games with a average user rating of 2.5")
```

```python
paid = df.loc[df['Price']> 0]
m.plot(paid['Average User Rating'] ,paid['Price'], 'og')
m.show()
print("Inference/observation:\nThe given graph shows us an even distribution where there is a hike in price mostly for the apps with user ratings ranging from 3.5 to 4.5")
```
