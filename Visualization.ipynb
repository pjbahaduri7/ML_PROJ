```python
# The two entirely different datasets being extracted and mined to show the use of various data visualization techniques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# This is a iris dataset basically dealing with flower parts
iris = pd.read_csv(r"C:\Users\pjbahaduri7\Downloads\iris.data", names=["Sep_len", "Sep_wid", "Pet_len", "Pet_wid", "class"])
iris
```

```python
# This dataset deals with the different wine types and the taster information
wine = pd.read_csv(r"C:\Users\pjbahaduri7\Desktop\wine_review.csv")
wine
```

```python
# representation of scatterplot 
fig, ax = plt.subplots()
ax.scatter(iris["Sep_len"], iris["Sep_wid"])
ax.set_title("Iris Dataset")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")

# using pandas visuaization
iris.plot.scatter(x="Sep_len", y="Sep_wid", title="Iris Dataset")

# using seaborn visualization
sns.scatterplot(x="Sep_len", y="Sep_wid", hue="class", data=iris)
```

```python
colors = {"Iris-setosa":"r", "Iris-versicolor":"g", "Iris-virginica":"b"}
fig, ax = plt.subplots()
for i in range(len(iris["Sep_len"])):
    ax.scatter(iris["Sep_len"][i], iris["Sep_wid"][i], c=colors[iris['class'][i]])
ax.set_title("Iris Dataset")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
```

```python
# Line chart representation 
# using pandas visualization
iris.drop(["class"], axis=1).plot.line(title="Iris Dataset")

# using seaborn visualization
sns.lineplot(data=iris.drop(["class"], axis=1))
```

```python
# Histogram representation to review the scores of different wines from dataset
fig, ax = plt.subplots()
ax.hist(wine["points"])
ax.set_title("Wine dataset Review scores")
ax.set_xlabel("Points")
ax.set_ylabel("Frequency")

# using pandas visualization
wine["points"].plot.hist()

# using seaborn visualization
# Gaussian KDE-Kernel Density Estimate
sns.distplot(wine["points"], bins=10, kde=True)
```

```python
# using bar graph visualization
fig, ax = plt.subplots()
data = wine["points"].value_counts()
points = data.index
frequency = data.values
ax.bar(points, frequency)
ax.set_title("Wine dataset Review scores")
ax.set_xlabel("Points")
ax.set_ylabel("Frequency")

# using pandas visualization
wine["points"].value_counts().sort_index().plot.bar()

wine.groupby("country").price.mean().sort_values(ascending=True)[:5].plot.bar()

# using seaborn visualization
sns.countplot(wine["points"])
```

```python
# box plot visualization which is used to display 5 number summary
df = wine[(wine["points"]>=95) & (wine["price"]<1000)]
sns.boxplot("points", "price", data=df)
```

```python
# Heatmap visualization is a graphical representation of data where the individual values contained within the matrix are 
# annoted with the help of colors and this is specially used for exploring the correlation of features in a dataset 
cor = iris.corr()
fig, ax = plt.subplots()
hm = ax.imshow(cor.values)
ax.set_xticks(np.arange(len(cor.columns)))
ax.set_yticks(np.arange(len(cor.columns)))
ax.set_xticklabels(cor.columns)
ax.set_yticklabels(cor.columns)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
# to show annotations of values in the heatmap use two for loops
# for i in range(len(cor.columns)):
#     for j in range(len(cor.columns)):
#         text = ax.text(j, i, np.around(cor.iloc[i, j], decimals=2), ha="center", va="center", color="red")

# using seaborn
sns.heatmap(iris.corr(), annot=True)
```

```python
# Faceting is a act of breaking the data variable across multiple subplots and combining them into a single figure
g = sns.FacetGrid(iris, col="class")
g = g.map(sns.kdeplot, "Pet_len")
```
